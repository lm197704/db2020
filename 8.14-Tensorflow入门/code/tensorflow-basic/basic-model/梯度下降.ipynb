```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda x : (x - 2.5)**2 + 5
# f(x) = (x - 2.5)**2 + 5
# t = 2(x - 2.5)
```

```python
x = np.linspace(0,5,100)

y = f(x)

plt.plot(x,y)
```

```python
# 导数
t = lambda x : 2*x - 5
```

```python
f(0.5)
```

```python
# 梯度下降求解最小，最快（梯度是导数，最陡峭，最快）

step = 0.1

# 退出条件，精确度
precision = 1e-5

# 初始值
s = 0.5

s_old = 0

# 将每次梯度下降求解的答案保存起来
ret_ = []

while True:
    
    print(s)
    s_old = s
    ret_.append(s)
    
    s = s - step*t(s)
    
    if np.abs(s  -s_old) < precision:
        break
```

```python
x = np.linspace(0,5,100)
plt.figure(figsize=(12,9))
plt.plot(x,f(x))

ret = np.array(ret_)

plt.scatter(ret,f(ret),color = 'red',marker = '+')
```
