```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

```

图片灰度处理

```python
plt.figure(figsize=(10,8))
jin=plt.imread('../data/jinzhengen.png')
plt.imshow(jin)
plt.show()

```

```python
#最大值法\最小值\平均值
print(jin.shape)
jin_max=jin.max(axis=2)
jin_max.shape
plt.imshow(jin_max)
plt.show()
plt.imshow(jin_max,cmap='gray')
plt.show()
```

```python
#加权平均值法RGB,最接近人眼的感觉
jin_weight=np.dot(jin,[0.299,0.587,0.114])/3
plt.imshow(jin_weight,cmap='gray')
plt.show()
```

```python
#matplotlib基础
#单一曲线
n=np.array(np.random.randint(0,10,size=10))
#plt.plot只能画线型图(x轴数据，y轴数据),x轴数据不必须
#其他：plt.bar\plt.hist\plt.scatter
plt.plot(np.arange(10),n)
plt.show()
plt.plot(n)
plt.show()
```

```python
#多条曲线
plt.plot(n)
plt.plot(n,n)
plt.show()
```

```python
#网格线及风格
plt.plot(n,n)
plt.grid(lw=2,color='red',linestyle='--',alpha=0.3)
```

```python
#练习：画三个子图在一个plt里面
plt.figure(figsize=(3*6,6))
axis1=plt.subplot(1,3,1)
x=np.linspace(-20,20,1000)
y=np.sin(x)
axis1.plot(x,y)
axis1.grid(lw=2)

axis2=plt.subplot(1,3,2)
axis2.plot(x,np.cos(x))
axis2.grid(c='r',alpha=0.4)

axis3=plt.subplot(1,3,3)
axis3.plot(x,np.power(x,2))
axis3.grid(c='green',linestyle='-.')
```

```python
#另外一种画法
figure=plt.figure(figsize=(3*6,6))
axis1=figure.add_subplot(1,3,1)
axis1.plot(x,y)
axis1.grid()

```

```python
#控制坐标轴界限
x=np.linspace(-1,1,100)
y=(1-x**2)**0.5
plt.plot(x,y)
#plt.show()
plt.plot(x,y,x,-y)
#plt.show()

#显示\设置当前坐标界限
plt.axis([-1,1,-1,1])
plt.axis(xmin=-1,xmax=3)
# plt.axis('xxx') 'tight'、'off'、'equal'……
plt.axis('equal')
plt.xlim(xmin=-2,xmax=3)
```

```python
#坐标轴标签\图标签
plt.plot(x,y,x,-y)
plt.axis('equal')
plt.xlabel('X',fontdict=dict(fontsize=20))
plt.ylabel('y=(1-x**2)**0.5',rotation=45,position=(0,0))
plt.title('title')
plt.show()
#图例
plt.plot(x,y,label='up')
plt.plot(x,-y,label='down')
#图例参数loc
plt.legend(loc=6)
plt.show()
```
