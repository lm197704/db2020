```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fftpack import fft2
%matplotlib inline
```

```python
moon=plt.imread('../data/moonlanding.png')
plt.figure(figsize=(10,8))
plt.imshow(moon,cmap='gray')
plt.show()
```

```python
moon_fft=fft2(moon)
moon_fft
```

```python
#>800
moon_fft[np.abs(moon_fft)>4e2]=0
```

```python
#
result=np.where(np.abs(moon_fft)>8e2,0,moon_fft)
#
from scipy.fftpack import ifft2
imoon_fft=ifft2(moon_fft)
moon_clean=np.real(imoon_fft)
```

```python
plt.figure(figsize=(10,8))
plt.imshow(moon_clean,cmap='gray')
```
